Для начала импортируем *poisson, binom, norm* из модуля scipy.stats.  

```python
from scipy.stats import poisson, binom, norm
```

Для того, чтобы получить случайную выборку из заданного распределения, мы можем воспользоваться функцией *rvs*. Проиллюстрируем на примере распределения Пуассона. Первый параметр, который мы должны указать в скобках, это лямбда ($\lambda$) для распределения Пуассона (среднее количество событий за заданный промежуток времени), второй - size - размер выборки. К примеру, если мат. ожидание = 3 и количество наблюдений = 10, то

```python
poisson.rvs(mu = 3, size = 10)
```

Для того, чтобы посчитать вероятность получить определенное значение сл.в. (количество событий за заданный промежуток времени в случае распределения Пуассона), мы используем функцию *pmf*, указав в скобках $k$ - значение сл.в., а также $\lambda$.

Давайте разберем на примере задачи из второго семинарского листка:

*Количество голов, забитой футбольной командой N соперникам, имеет распределение Пуассона. Известно, что в среднем за один тайм (45 минут) команда N забивает 2 гола. Найдите вероятность того, что за футбольный матч (90 минут) команда N
забьет 3 мяча?*

```python
poisson.pmf(k = 3, mu = 4)
```

Если же нас интересует вероятность того, что за 90 минут команда N забьет менее 3 мячей, то можно представить это как функцию распределения от 2. Для расчета $F(2)$ воспользуемся функцией *cdf*.

```python
poisson.cdf(k = 2, mu = 4)
```

Для того, чтобы посчитать вероятность того, что сл.в. превышает заданное значение, в Python существует функция *sf* (аббревиатура от survival function).

*В среднем на 4 страницы текста книги приходится 1 опечатка. Найдите вероятность того, что на одной странице текста книги встретится более одной опечатки, если сл. в. «число опечаток в книге» имеет приближенно распределение Пуассона?*

```python
poisson.sf(k = 1, mu = 0.25)
```

*Вероятность того, что пассажир не придет к отправлению автобуса, равна 0.01. Найдите вероятность того, что 2 из 30 пассажиров не придут к
отправлению автобуса. Сравните ответы, полученные с помощью биномиального

В домашнем задании Вы довольно много тренировались находить вероятности и квантили по таблице стандартного нормального распределения. Теперь сделаем это в Python автоматически

Итак, найдем $P(Z\leq 1)$, если $Z\sim N(0,1)$. По умолчанию в norm функции параметры устанавливаются для стандартного нормального распределения.

```python
norm.cdf(1)
```

Что если нам нужно найти вероятность для произвольного нормального распределения? К примеру, $P(X\geq 6)$, если $X\sim N(5;1)$

```python
norm.sf(6,5,1) #здесь последовательно указаны значение сл.в., мат.ожидание и стандартное отклонение произвольного нормального распределения
```

Найдем квантиль для стандартного нормального распределения уровня 0.995

```python
norm.ppf(0.995)
```
